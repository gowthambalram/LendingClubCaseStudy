# Lending Data Analysis Notebook
## This section provides a preliminary look at the data columns and initial statistics.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load data
loan_data = pd.read_csv('loan.csv')
# Display head
loan_data.head()
```

```python
# Analyse Data in loan.csv
loan_data.info()
```

```python
# Describe to understand various columns
loan_data.describe()
```

# Data Cleaning
## This section will address missing values, duplicates, and data anomalies.

```python
# Calculate missing values percentages
missing_values = loan_data.isnull().mean() * 100
missing_values[missing_values > 0].sort_values(ascending=False)
```

## Removing Columns with High Missing Values. Columns with more than 50% missing data are considered less reliable and is removed in the following section

```python
# Removing columns with high missing values
columns_to_remove = missing_values[missing_values > 50].index
loan_data.drop(columns=columns_to_remove, inplace=True)
loan_data.shape
```

```python
# Remove duplicates if any
duplicates = loan_data.duplicated().sum()
if duplicates > 0:
    loan_data.drop_duplicates(inplace=True)
loan_data.shape
```

# Univariate Analysis
## This section examines individual variables to understand their distributions and identify any anomalies or outliers.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Analyzing the distribution of loan amounts
plt.figure(figsize=(10, 6))
sns.histplot(loan_data['loan_amnt'], bins=30, kde=False, color='blue')
plt.title('Loan Amount Distribution')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyzing the distribution of interest rates
plt.figure(figsize=(10, 6))
sns.histplot(loan_data['int_rate'].str.rstrip('%').astype('float'), bins=30, kde=False, color='green')
plt.title('Interest Rate Distribution')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

- **Univariate Analysis Insights**: The analysis identified that the distribution of loan amounts showed multiple peaks, typically around $5000, $10,000, $13,000, and $15,000, indicating popular loan amounts. Interest rates mostly clustered around 10% to 15%, highlighting common lending rates.

```python
# Analyzing the distribution of loan statuses
plt.figure(figsize=(10, 6))
loan_status_counts = loan_data['loan_status'].value_counts()
sns.barplot(x=loan_status_counts.index, y=loan_status_counts.values, palette='viridis')
plt.title('Loan Status Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# Segmented Univariate Analysis
## This section explores how key financial metrics vary across different segments, specifically focusing on the loan purpose.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Segmenting loan amount and interest rate by purpose
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.boxplot(x='purpose', y='loan_amnt', data=loan_data)
plt.title('Loan Amount by Purpose')
plt.xticks(rotation=90)
plt.xlabel('Purpose')
plt.ylabel('Loan Amount')
plt.subplot(1, 2, 2)
sns.boxplot(x='purpose', y='int_rate', data=loan_data)
plt.title('Interest Rate by Purpose')
plt.xticks(rotation=90)
plt.xlabel('Purpose')
plt.ylabel('Interest Rate (%)')
plt.tight_layout()
plt.show()
```

- **Segmented Univariate Analysis**: Segmented analysis by loan purpose showed that loans for small business, debt consolidation and house had higher average amounts and rates, identifying them as high-risk areas. This segmentation helps in tailoring risk management strategies for different loan purposes.

# Bivariate Analysis
## This section explores relationships between pairs of variables to understand how they interact with each other.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'int_rate' from a string percentage to a float
loan_data['int_rate'] = loan_data['int_rate'].str.rstrip('%').astype(float)

# Analyzing relationship between loan amount and interest rate
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.scatterplot(x='loan_amnt', y='int_rate', data=loan_data, alpha=0.5)
plt.title('Loan Amount vs. Interest Rate')
plt.xlabel('Loan Amount')
plt.ylabel('Interest Rate (%)')
# Analyzing interest rate distribution by loan status
plt.subplot(1, 2, 2)
sns.boxplot(x='loan_status', y='int_rate', data=loan_data)
plt.title('Interest Rate by Loan Status')
plt.xlabel('Loan Status')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Calculating correlation between loan amount and interest rate
loan_data[['loan_amnt', 'int_rate']].corr()
```

- **Bivariate Analysis Insights**: The relationship between loan amount and interest rate revealed a moderate positive correlation, suggesting that higher loan amounts generally come with higher interest rates due to increased risk. This insight is vital for risk-based pricing strategies.

# Correlation Analysis
## This section displays the correlation matrix for key variables to identify relationships and potential collinearity.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Correlation matrix among key variables
key_variables = ['loan_amnt', 'int_rate', 'annual_inc', 'dti']
correlation_matrix = loan_data[key_variables].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Key Variables')
plt.show()
```

# Time Series Analysis
## This section analyzes trends over time for key financial metrics such as average loan amounts and interest rates.

```python
import matplotlib.pyplot as plt
# Setting index for time series analysis
loan_data['issue_d'] = pd.to_datetime(loan_data['issue_d'], format='%b-%y')
time_data = loan_data.set_index('issue_d').resample('M').agg({'loan_amnt': 'mean', 'int_rate': 'mean'})
# Plotting trends over time
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
time_data['loan_amnt'].plot(title='Trend of Average Loan Amounts Over Time', color='blue')
plt.xlabel('Year')
plt.ylabel('Average Loan Amount')
plt.subplot(1, 2, 2)
time_data['int_rate'].plot(title='Interest Rates Over Time', color='green')
plt.xlabel('Year')
plt.ylabel('Average Interest Rate (%)')
plt.tight_layout()
plt.show()
```

# Default Rate Analysis
## This section examines the default rates across different loan grades and purposes to identify risk factors.

```python
import matplotlib.pyplot as plt
# Calculating default rates by loan grade and purpose
loan_data['default'] = loan_data['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)
default_rate_grade = loan_data.groupby('grade')['default'].mean()
default_rate_purpose = loan_data.groupby('purpose')['default'].mean()
# Plotting default rates
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
default_rate_grade.plot(kind='bar', color='coral')
plt.title('Default Rate by Loan Grade')
plt.xlabel('Loan Grade')
plt.ylabel('Default Rate')
plt.subplot(1, 2, 2)
default_rate_purpose.plot(kind='bar', color='teal')
plt.title('Default Rate by Loan Purpose')
plt.xlabel('Loan Purpose')
plt.ylabel('Default Rate')
plt.tight_layout()
plt.show()
```

# Yearly Default Trends Analysis
## This section analyzes the default rates on a yearly basis to observe trends over time.

```python
import matplotlib.pyplot as plt
# Yearly default rate calculation
yearly_default_trends = loan_data.resample('A', on='issue_d')['default'].mean()
# Plotting yearly default trends
plt.figure(figsize=(10, 6))
yearly_default_trends.plot(kind='line', marker='o', color='red', linestyle='-')
plt.title('Yearly Default Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Default Rate')
plt.grid(True)
plt.show()
```

## Additional Analysis on Default Loans

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create 'interest_rate_segment' by categorizing 'int_rate'
loan_data['interest_rate_segment'] = pd.cut(loan_data['int_rate'], bins=[0, 5, 10, 15, 20, 25, 30], labels=['0-5%', '5-10%', '10-15%', '15-20%', '20-25%', '25-30%'])

# Plotting Default Loans by Interest Rate Segments
plt.figure(figsize=(10, 6))
sns.countplot(x='interest_rate_segment', data=loan_data[loan_data['default'] == 1], palette='viridis')
plt.title('Default Loans Distribution by Interest Rate Segments')
plt.xlabel('Interest Rate Segments')
plt.ylabel('Number of Defaults')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- **Insight for 'Default Loans Distribution by Interest Rate Segments':** Higher default rates are apparent in higher interest rate segments.

- **Interest Rate and Defaults**: Analysis shows that default rates increase significantly with higher interest rate segments, particularly in the 10-15% and 15-20% brackets.

- **Loan Amount and Defaults**: Larger loan amounts correlate with higher default rates, suggesting a need for stricter lending criteria for larger loans.

## Analysis of Default Rates by State

## Heatmap of Average Loan Amounts by Income Bracket and State

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert annual_inc to a categorical variable based on income brackets
loan_data['income_bracket'] = pd.cut(loan_data['annual_inc'],
                                     bins=[0, 25000, 50000, 75000, 100000, 125000, 150000, float('inf')],
                                     labels=['<25k', '25k-50k', '50k-75k', '75k-100k', '100k-125k', '125k-150k', '>150k'])

# Create a pivot table to analyze average loan amounts by income bracket and state
state_income_loan_amount = loan_data.pivot_table(values='loan_amnt', index='addr_state', columns='income_bracket', aggfunc='mean')


# Heatmap of loan amounts by income and state
plt.figure(figsize=(16, 12))
sns.heatmap(state_income_loan_amount, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=.5)
plt.title('Average Loan Amount by Income Bracket and State')
plt.xlabel('Income Bracket')
plt.ylabel('State')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- **Insight for 'Average Loan Amount by Income Bracket and State':** Highlights state-level and income-based variations in average loan amounts.

## Heatmap of Employment Length Distribution by State

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'emp_length' to string to ensure the .str accessor works
loan_data['emp_length'] = loan_data['emp_length'].astype(str)

# Now, extract numbers from the 'emp_length' string
loan_data['emp_length'] = loan_data['emp_length'].str.extract(r'(\d+)').astype(float)  # Use raw string for regex

# Directly handle NaN values in the DataFrame to avoid the FutureWarning
loan_data['emp_length'] = loan_data['emp_length'].fillna(0)  # Replace NaNs with 0 directly

# Create a pivot table to analyze employment length distribution by state
state_employment_trends = loan_data.pivot_table(index='addr_state', columns='emp_length', values='id', aggfunc='count')

# Heatmap of employment lengths by state
plt.figure(figsize=(18, 14))
sns.heatmap(state_employment_trends, annot=False, cmap='BuPu', linewidths=.5)
plt.title('Employment Length Distribution by State')
plt.xlabel('Employment Length')
plt.ylabel('State')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- **Insight for 'Employment Length Distribution by State':** Displays varying employment stability across states.

## Additional Default Trend Visualizations

```python
import matplotlib.pyplot as plt
# Default Rate by Year

# Convert 'issue_d' to datetime, and extract the year
loan_data['issue_d'] = pd.to_datetime(loan_data['issue_d'])
loan_data['issue_year'] = loan_data['issue_d'].dt.year

# Calculate the default rate by year
# Assuming 'default' is a column where 1 indicates a default and 0 indicates no default
default_rate_by_year = loan_data.groupby('issue_year')['default'].mean()

plt.figure(figsize=(12, 6))
default_rate_by_year.plot(kind='line', marker='o', color='red')
plt.title('Default Rate by Issue Year')
plt.xlabel('Year')
plt.ylabel('Default Rate')
plt.grid(True)
plt.tight_layout()
plt.show()
```

- **Insight for 'Default Rate by Issue Year':** Indicates fluctuating default rates over the years, reflecting economic or policy changes.

```python
# Default Rate by Loan Grade

# Calculate the default rate by loan grade
default_rate_by_grade = loan_data.groupby('grade')['default'].mean()

plt.figure(figsize=(12, 6))
default_rate_by_grade.sort_values().plot(kind='bar', color='blue')
plt.title('Default Rate by Loan Grade')
plt.xlabel('Loan Grade')
plt.ylabel('Default Rate')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

- **Insight for 'Default Rate by Loan Grade':** Illustrates higher default rates in higher grades, affirming the risk associated with these grades.

## Analysis of Defaults by Financial Maturity

```python
import matplotlib.pyplot as plt
# Plotting default rate by financial maturity

# Specify the exact format of your date data if known, for example 'MM/YYYY'
loan_data['earliest_cr_line'] = pd.to_datetime(loan_data['earliest_cr_line'], format='%m/%Y', errors='coerce')
loan_data['credit_start_year'] = loan_data['earliest_cr_line'].dt.year

# Calculate financial maturity as the difference from the current year
current_year = pd.Timestamp('now').year
loan_data['financial_maturity'] = current_year - loan_data['credit_start_year']

# Handle NaN values by assigning directly to the column
loan_data['financial_maturity'] = loan_data['financial_maturity'].fillna(0)  # Replace NaNs with 0 directly

# Calculate the default rate by financial maturity
default_rate_by_financial_maturity = loan_data.groupby('financial_maturity')['default'].mean()

plt.figure(figsize=(12, 6))
default_rate_by_financial_maturity.plot(kind='bar', color='teal')
plt.title('Default Rate by Financial Maturity (Years)')
plt.xlabel('Years of Financial Maturity')
plt.ylabel('Default Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
